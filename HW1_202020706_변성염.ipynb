### 문제 2-2
오늘 날짜는 2024년 3월 16일 입니다.

`Year_Birth` 열은 현재 **년도** 단위로 데이터가 구성되어 있습니다.

**나이** 단위로 데이터를 변경하고 캡처하여 보고서에 업로드 하세요.

예)  1957년도 출생 => 2024-1957=67

`DT_Customer` 열은 현재 **날짜** 단위로 데이터가 구성되어 있습니다.  

**년** 단위로 데이터를 변경하고 캡처하여 보고서에 업로드 하세요.

예) 04-09-2012 => (03-16-2024-04-09-2012)/365=11.94 (소수점 셋째자리에서 버림)

**Hint) datetime.strptime**

```python
from datetime import datetime
################  문제 2-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################
format = "%d-%m-%Y" # strptime 사용을 위한 기존 데이터의 날짜 형식

def getYears(date): # 연도/월/일 형식의 날짜를 년 단위로 변경하는 함수
  dateobj = datetime.strptime(date, format) # 년 단위 변경의 기준이 되는 날짜의 datetime 객체 생성
  result = int((datetime.strptime("16-03-2024", format) - dateobj).days / 365 * 100) / 100 # 현재 날짜(2024-03-16)와의 차이를 연 단위로 계산 후 소수점 2자리까지만
  return result

data['Year_Birth'] = 2024 - data['Year_Birth'] # Year_Birth 열의 데이터들을 2024 - 기존 값, 즉 '나이'로 단위 변경
data['Dt_Customer'] = data['Dt_Customer'].apply(getYears) # apply()를 사용해 데이터들에 getYears함수 적용
################ 코드 끝 ########################
#################################################
```

### 문제 3-2
주어진 코드틀에 k개의 "MntToyProducts" 값을 보고 평균을 내어 결측값을 결정하는 코드를 작성하세요.
* 주의) 코드틀 이외의 부분에 코드를 작성하지 마세요!
* 주의) 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
* 주의) 코드 시작과 코드 끝 라인을 삭제하지 마세요.

```python
from tqdm.notebook import tqdm
```

```python
class Knn_imputer():
  def __init__(self, n_neighbors):

    ## 생성자로 최근접 이웃수(n_neighbors)를 입력 받습니다.
    self.k = n_neighbors

  ## knn 은 lazy learning 입니다.
  def fit(self,X):
      self.X = X

  def predict(self, X):

    for i in tqdm(range(len(X))):
     if pd.isnull(X['MntToyProducts'].iloc[i]): # Arrival Delay 열에 있는 결측값 확인

        distance = self.dist(X, X.iloc[i]) # 데이터프레임 X와 결측값이 있는 행 데이터 간의 거리 구하는 함수
        sorted_dist = sorted(distance.items(), key=lambda x: x[1]) # distance 를 sorting 하기

         ################  문제 3-2  ####################
         # k개의 "MntToyProducts" 값을 보고 평균 내어
         # 값을 결정하기

         # 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
         # 코드 시작과 코드 끝 라인을 삭제하지 마세요.
        ################ 코드 시작  #####################
        # 가장 가까운 k개의 distance를 더한 뒤 평균을 구하는 코드
        temp = 0
        for j in range(self.k):
          if sorted_dist[j][0] != i: # 자기 자신은 고려하지 않기 위해
            temp += X['MntToyProducts'].iloc[sorted_dist[j][0]] # 가장 가까운 Kidhome 행들의 MntToyProducts값을 고려
          else:
            j -= 1
        temp /= self.k

        X['MntToyProducts'].iloc[i] = temp # 구한 값을 결측값 대신 대입
        ################ 코드 끝  ######################
        ##############################################

    return X['MntToyProducts']

  def dist(self, X, y):
    num_train = X.shape[0]

    X = X.drop(['MntToyProducts'], axis=1)
    y = y.drop(['MntToyProducts'])
    dist_dict = {} ## {index : distance} 형태

    for i in range(num_train):

    ################  문제 3-1  ####################
    # 데이터프레임 X와 결측값이 있는 행 데이터 간의 거리 구하기
    # 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
    # 코드 시작과 코드 끝 라인을 삭제하지 마세요.
    ################ 코드 시작  #####################
      dist_dict[i] = (y[0] - X.iloc[i][0]) * (y[0] - X.iloc[i][0])
      # 결측값이 있는 행의 Kidhome 값과 다른 행의 Kidhome 사이의 거리들을 저장(행 정보 i를 사전의 key로 저장), 방향을 고려하지 않으므로 제곱
    ################ 코드 끝  ######################
    ##############################################

    return dist_dict
```

```python
imputer = Knn_imputer(3)
imputer.fit(X)

# test set에 대해 predict를 진행합니다.
data['MntToyProducts'] = imputer.predict(X) # 오래 걸립니다
```

### 문제 5-1
**`KNeighborsClassifier`**를 사용하여 주어진 조건에 맞게 모델을 학습하고 완성한 코드를 보고서에 캡처하여 업로드하세요.
* 조건 1: neighbor의 수 `k= [1, 3, 5, 15, 55, 101, 501]`으로 변화에 따른 정확도의 변화를
`line plot`으로 그리고 가장 높은 성능의 `k`를 출력하시오.
* 조건 2: train set에서 validation set을 나누세요. `train : validation = 7 : 3`

* *주의! : test set이 아닌 validation set으로 평가해야 합니다.*

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
################  문제 5-1  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################
numK = (1, 3, 5, 15, 55, 101, 501) # K로 사용할 값들

# 데이터를 train set과 Validation set으로 다시 분할
X_v_train, X_v_Test, y_v_train, y_v_test = train_test_split(X_train, y_train, test_size = 0.3)
knn = KNeighborsClassifier() # knn classifier 생성
knn_scores = [] # 점수(정확도)를 저장할 배열

for i in range (7): # 각 K에 대해
  knn.n_neighbors = numK[i]
  knn.fit(X_v_train, y_v_train) # 학습 진행
  knn_scores.append(knn.score(X_v_Test, y_v_test)) # 성능 평가 및 점수 저장

plt.plot([x for x in range(1, 8)], knn_scores) # 간격을 동일하게 하기 위해 x축을 1~7의 정수값으로 설정
plt.xticks(ticks = [x for x in range(1, 8)], labels= [str(x) for x in numK]) # x값들을 원래의 값들과 매핑
plt.xlabel('n of neighbor')
plt.ylabel('Accuracy')

print("Best k : ", numK[knn_scores.index(max(knn_scores))]) # 최고의 K(가장 높은 정확도가 나오는 시점의 K)
################ 코드 끝  ######################
##############################################
```

```python
sns.lineplot(x = numK, y = knn_scores)
```

### 문제 5-2
**`DecisionTreeClassifier`**를 사용하여 주어진 조건에 맞게 모델을 학습하고 완성한 코드 및 출력물을 보고서에 캡처하여 업로드하세요.
* 조건 1: 최대 트리의 깊이를 `depth  = [1, 3, 5, 10, 15, 30, 20]`으로 변화에 따른 정확도의 변화를 `line plot`으로 그리고 가장 높은 성능의 `depth`를 출력하시오.
* 조건 2: train set에서 validation set을 나누세요. `train : validation = 7 : 3`

* *주의! : test set이 아닌 validation set으로 평가해야 합니다.*

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
################  문제 5-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################
num_dep = (1, 3, 5, 10, 15, 30, 20) # depth로 사용할 값들

# 데이터를 train set과 Validation set으로 다시 분할
X_v_train, X_v_Test, y_v_train, y_v_test = train_test_split(X_train, y_train, test_size = 0.3)
DT = DecisionTreeClassifier()
dt_scores = [] # 점수(정확도)를 저장할 배열

for i in range (7): # 각 depth에 대해
  DT.max_depth = num_dep[i]
  DT.fit(X_v_train, y_v_train) # 학습 진행
  dt_scores.append(DT.score(X_v_Test, y_v_test)) # 성능 평가 및 점수 저장

plt.plot([x for x in range(1, 8)], dt_scores) # 간격을 동일하게 하기 위해 x축을 1~7의 정수값으로 설정
plt.xticks(ticks = [x for x in range(1, 8)], labels= [str(x) for x in numK]) # x값들을 원래의 값들과 매핑
plt.xlabel('depth')
plt.ylabel('Accuracy')

print("Best depth : ", num_dep[dt_scores.index(max(dt_scores))]) # 최고의 K(가장 높은 정확도가 나오는 시점의 K)
################ 코드 끝  ######################
##############################################
```

```python
sns.lineplot(x = num_dep, y = dt_scores)
```

```python
################  문제 5-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# 위에서 구한 최고의 k와 depth값 입력
best_k = 15
best_depth = 5

# 각 분류기의 hyperparameter 수정
knn.n_neighbors = best_k
DT.max_depth = best_depth

# 전체 학습 셋으로 학습 진행
knn.fit(X_train, y_train)
DT.fit(X_train, y_train)

# 평가 셋으로 정확도 측정
Knn_acc = knn.score(X_test, y_test)
DT_acc = DT.score(X_test, y_test)

print(Knn_acc)
print(DT_acc)
################ 코드 끝  ######################
##############################################
```
